#1. Data Collection

The data obtained in this study was internal data extracted from the company's issue tracking software. The data owner is the IT Support Team and information recorded is based on all projects from a few of the operation systems (application). Dataset supplied spanned 4.5 years, from 29 October 2016 to 12 May 2021. This provides us with 17 months of data after the new update launched at 1 December 2019. 

There had been a total of 304 projects and 176,299 issues logged throughout the given time period. There are 10 variables extracted to be used for this current study, further information pertaining to the variables can be found in the data dictionary under Table 1.

To obtain more insights, Graph 4 was plotted based on a number of issues at the country level. The illustration demonstrates Mainland China having the highest number of issues which peaked at 3,567 issues in June 2018. Following new update, there is a notable drop after December 2019. 
Hong Kong, on the other hand, had a smaller degree of fluctuations till the year 2020 whereby the number of issues spikes to 1,811 issues in November 2020. Thereafter, the number of issues declined.
The remaining countries had minimal fluctuations as the majority of the operation were in either Mainland China or Hong Kong.
